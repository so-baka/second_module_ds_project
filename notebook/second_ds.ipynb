После этого добавьте два бинарных признака:

- `is_two` — совершил ли пользователь 2 и более заказа;
- `is_five` — совершил ли пользователь 5 и более заказов.

**Рекомендация:** перед тем как строить профиль, отсортируйте данные по времени совершения заказа.

---


```python
df = df.sort_values(['user_id', 'order_ts']) #сортируем по времени совершения заказа

user_profile = df.groupby('user_id').agg( #агрегация всех нужных показателей с группировкой по пользователям
    first_order_dt=('order_dt', 'first'),
    last_order_dt=('order_dt', 'last'),
    first_device=('device_type_canonical', 'first'),
    first_region=('region_name', 'first'),
    first_partner=('service_name', 'first'),
    first_genre=('event_type_main', 'first'),
    total_orders=('order_id', 'count'),
    avg_revenue_rub=('revenue_rub', 'mean'),
    avg_tickets=('tickets_count', 'mean')
).reset_index()

avg_days_between = df.groupby('user_id')['days_since_prev'].mean().reset_index()
avg_days_between.columns = ['user_id', 'avg_days_between']

user_profile = pd.merge(user_profile, avg_days_between, on='user_id', how='left')

user_profile['is_two'] = (user_profile['total_orders'] >= 2).astype(int)
user_profile['is_five'] = (user_profile['total_orders'] >= 5).astype(int)


print(user_profile.head())
print(f"\nВсего пользователей: {len(user_profile)}")
print(f"Совершили >=2 заказов: {user_profile['is_two'].sum()}")
print(f"Совершили >=5 заказов: {user_profile['is_five'].sum()}")
```

```python
na_count = user_profile['avg_days_between'].isna().sum()
total_count = len(user_profile)
na_percent = (na_count / total_count * 100).round(2)
print(f"Пропусков в поле avg_days_between: {na_count} ({na_percent}%)")
```

Для каждого пользователя был найден требуемый показатель и созданы два новых бинарных поля: столбец `is_two`, указывающий на то, что пользователь совершил 2 и более заказов и столбец `is_five`, говорящий о том, что пользователь совершил 5 и более заказов. Первых 13559 человек, вторых 6400. Появившиеся пропуски в поле `avg_days_between` оставлены как есть - заполнение их числовым индикатором нарушит статистические показатели по полю, наличие пропусков не помешает дальнейшим расчетам и корреляции, в частности. Удаление пропусков так же нецелесообразно - может затронуть важные данные в других полях.

При интерпретации результатов учитывайте размер сегментов: если в сегменте мало пользователей (например, десятки), то доли могут быть нестабильными и недостоверными, то есть показывать широкую вариацию значений.

```python
# Вычисляем среднюю долю повторных пользователей по всей выборке
avg_retention = user_profile['is_two'].mean() * 100
print(f"Средняя доля повторных пользователей по всей выборке: {avg_retention:.2f}%\n")
```

```python
# функция для анализа сегмента
def analyze_segment_return(segment_col, top_n=10):
    segment_stats = user_profile.groupby(segment_col).agg( # Группируем по сегменту

        users_count=('user_id', 'count'),
        retention_rate=('is_two', 'mean')
    ).reset_index()
    
    segment_stats['retention_rate'] = (segment_stats['retention_rate'] * 100).round(2)# доля в процентах
    segment_stats = segment_stats.sort_values('users_count', ascending=False)
    
    top_segments = segment_stats.head(top_n).copy()# топ-n по количеству пользователей
    
    top_segments['reliable'] = top_segments['users_count'] >= 100  # порог можно менять
    
    return top_segments
```

```python
# по типу первого мероприятия
genre_ret = analyze_segment_return('first_genre')
print("Возвраты по жанру первого мероприятия")
print(genre_ret[['first_genre', 'users_count', 'retention_rate', 'reliable']])
print()
```

Больший воврат у пользователей, посещающих концерты.

```python
# по типу устройства
device_ret = analyze_segment_return('first_device')
print("Возвраты по типу устройства")
print(device_ret[['first_device', 'users_count', 'retention_rate', 'reliable']])
print()
```

На второй заказ больше решаются пользователи стационарных устройств.

```python
# по региону (топ-10)
region_ret = analyze_segment_return('first_region', top_n=10)
print("Возвраты по региону (топ-10 по размеру)")
print(region_ret[['first_region', 'users_count', 'retention_rate', 'reliable']])
print()
```

Наиболее часто второй заказ совершают пользователи, посетившие свое первое мероприятие в Шанырском регионе.

```python
# по билетному оператору (топ-10)
partner_ret = analyze_segment_return('first_partner', top_n=10)
print("=== Возвраты по билетному оператору (топ-10) ===")
print(partner_ret[['first_partner', 'users_count', 'retention_rate', 'reliable']])
```

Наилучшие условия для второго раза предоставляет компания "Край билетов". Не исключено, что благодаря наилучшему первому разу.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

def plot_retention(ax, data, x_col, y_col, title, avg_line):
    bars = ax.bar(range(len(data)), data[y_col])
    ax.axhline(y=avg_line, color='red', linestyle='--', label=f'Среднее ({avg_line:.1f}%)')
    ax.set_xticks(range(len(data)))
    ax.set_xticklabels(data[x_col], rotation=45, ha='right')
    ax.set_ylabel('Доля повторных пользователей (%)')
    ax.set_title(title)
    ax.legend()
    
    for i, (bar, val) in enumerate(zip(bars, data[y_col])):
        if val > avg_line:
            bar.set_color('green')
        else:
            bar.set_color('gray')

plot_retention(axes[0,0], genre_ret, 'first_genre', 'retention_rate', 
               'Возвраты по жанру первого мероприятия', avg_retention)

plot_retention(axes[0,1], device_ret, 'first_device', 'retention_rate',
               'Возвраты по типу устройства', avg_retention)

plot_retention(axes[1,0], region_ret, 'first_region', 'retention_rate',
               'Топ-10 регионов по возвратам', avg_retention)

plot_retention(axes[1,1], partner_ret, 'first_partner', 'retention_rate',
               'Топ-10 операторов по возвратам', avg_retention)

plt.tight_layout()
plt.show()
```

Изучите, какие характеристики первого заказа и профиля пользователя могут быть связаны с числом покупок. Для этого используйте универсальный коэффициент корреляции `phi_k`, который позволяет анализировать как числовые, так и категориальные признаки.

**Задача 4.4.1:** Проведите корреляционный анализ:
- Рассчитайте коэффициент корреляции `phi_k` между признаками профиля пользователя и числом заказов (`total_orders`). При необходимости используйте параметр `interval_cols` для определения интервальных данных.
- Проанализируйте полученные результаты. Если полученные значения будут близки к нулю, проверьте разброс данных в `total_orders`. Такое возможно, когда в данных преобладает одно значение: в таком случае корреляционный анализ может показать отсутствие связей. Чтобы этого избежать, выделите сегменты пользователей по полю `total_orders`, а затем повторите корреляционный анализ. Выделите такие сегменты:
    - 1 заказ;
    - от 2 до 4 заказов;
    - от 5 и выше.
- Визуализируйте результат корреляции с помощью тепловой карты.

```python
cols_for_corr = [
    'avg_revenue_rub',
    'avg_tickets',
    'avg_days_between',
    'first_device',
    'first_genre',           # из корреляции исключаем total_orders, нет прикола смотреть на "1"
    'first_region',
    'first_partner',
    'first_weekday'
]

df_corr = user_profile[cols_for_corr + ['total_orders']]

interval_cols = ['avg_revenue_rub', 'avg_tickets', 'avg_days_between', 'first_weekday', 'total_orders'] #интервальные поля помечаем отдельно

phik_matrix = df_corr.phik_matrix(interval_cols=interval_cols)

corr_with_target = phik_matrix['total_orders'].drop('total_orders').sort_values(ascending=False)
print("=== Корреляция признаков с total_orders (phi_k) ===")
print(corr_with_target)
```

Как и ожидалось, к сожалению, phik считает корреляцию каждого показателя с каждым отдельным заказом, в следствие чего никаких взаимосвязей и быть не может. Самая сильная связь здесь 0.12 - это с натяжкой можно считать "слабой" корреляцией. Будем делить `total_orders` на три сегмента:
- 1 заказ
- 2-4 заказа
- 5 и более заказов

```python
user_profile['orders_segment'] = '1 заказ'
user_profile.loc[(user_profile['total_orders'] >= 2) & (user_profile['total_orders'] <= 4), 'orders_segment'] = '2-4 заказа'
user_profile.loc[user_profile['total_orders'] >= 5, 'orders_segment'] = '5+ заказов'

print(user_profile['orders_segment'].value_counts())
```

Взяли три сегмента из выборки, описанные выше. Теперь можно еще раз провести корреляционный анализ.

```python
feature_cols = [
    'avg_revenue_rub',
    'avg_tickets',
    'avg_days_between',
    'first_device',
    'first_genre',
    'first_region',
    'first_partner',
    'first_weekday'
]

interval_cols = ['avg_revenue_rub', 'avg_tickets', 'avg_days_between', 'first_weekday']

df_corr = user_profile[feature_cols + ['orders_segment']]

phik_matrix = df_corr.phik_matrix(interval_cols=interval_cols)

corr_with_segment = phik_matrix['orders_segment'].drop('orders_segment').sort_values(ascending=False)#убрали корреляцию с самим собой

print("Корреляция признаков с сегментом пользователя (orders_segment)")
print(corr_with_segment.round(3))
```

Теперь лучше. Видим умеренную корреляцию с полями `avg_days_between` и `avg_tickets`. Корреляция с `first_region` только-только попадает в категорию "слабая" корреляция, с остальными полями `total_orders` практически не имеет никакой связи. Это лучше будет видно на тепловой карте.

```python
corr_results = corr_with_segment.to_frame(name='phi_k')

plt.figure(figsize=(10, 8))
sns.heatmap(corr_results.T, annot=True, fmt='.3f', cmap='viridis', # очень уж мне понравился этот градиент
            cbar_kws={'label': 'Phi_k коэффициент'}, linewidths=1)
plt.title('Корреляция признаков с сегментом пользователя (orders_segment)', fontsize=14)
plt.xlabel('Признаки')
plt.ylabel('')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Тепловая карта подтверждает числовые данные - с количеством заказов наиболее связаны:
- Количество дней между заказами
- Среднее количество билетов в заказе
Как мы выяснили в прошлой задаче, чем меньше количетсво дней между заказами, тем большее количетсво заказов сделает человек, эта корреляция была установлена случайно еще там.
Пользователи, которые возвращаются редко, с высокой долей вероятности так и останутся в сегменте `2-4 заказа`.
На счет `avg_tickets`: Чем больше билетов за раз покупает пользователь, тем выше его итоговая активность. Это могут быть небольшие "корпоративы" или семейные меропрятия.

Наиболее близко к нему находится сегмент `3-5 билетов` с показателем повторного заказа в 54.91% и долей от общей выборки в 41.95%. `Стоит сосредоточить усилия маркетинга на пользователях, покупающих `2-5 билетов`, так мы получим долю в 85,91 от всей выборки, которые наиболее часто возвращаются к повторной покупке.`
- Присутвуют аномальные показатели:
    - Повторные заказы совершает 74.1% пользователей выборки из сегмента `2-3 билета`
    - Очень низкую долю выборки составляет сегмент `5+` - ироничные 3.14%. Можно полагать, что в этот сегмент входят "ценители" - пользователи, предпринимающие активную деятельность в культурном плане.
    
7. **Как день недели, в которой была совершена первая покупка, влияет на поведение пользователей**
- День недели практически не влияет на повторный заказ пользователем. Показатели сликом равномерно распределены, нельзя выделить какой-то конкретный день. `Воскресенье` проседает больше остальных по количетсву заказов - 12.85% против основного диапазона по остальным дням 13.44-15.78%. А `суббота` немного выделяется на фоне остальных по повторным заказам 64.11% против худшего дня повторных заказов - четверга с 59.58% повторов.

8. **Как средний интервал между заказами влияет на удержание клиентов**
- Для расчета брались два сегмента: пользователи, совершившие `2-4` покупки и пользователи, совершившие `5 и более` покупок
    - Пользователи сегмента `5+` совершают покупки билетов более чем в два раза чаще сегмента `2-4`.Разрыв между повторными заказами обратнопропорционален лояльности, то есть чем меньше дней прошло с предыдущего заказа, тем выше шанс, что пользователь совершит больше пяти покупок.

9. **Корреляционный анализ количества покупок и признаков пользователя**
- С количеством заказов наиболее связаны:
    - Количество дней между заказами
    - Среднее количество билетов в заказе
    - Чем меньше количетсво дней между заказами, тем большее количетсво заказов сделает человек.
    - Пользователи, которые возвращаются редко, с высокой долей вероятности так и останутся в сегменте `2-4 заказа`.
    - Чем больше билетов за раз покупает пользователь, тем выше его итоговая активность. Это могут быть небольшие "корпоративы" или семейные меропрятия.
